```python
'''
    1. Заполните недостающие части класса Calculator
       Заполните ____ пропуски в реализации Калькулятора, чтобы пройти утверждения.
'''

class Calculator:
    def __init__(self, var1, var2):
        self.var1 = var1
        self.var2 = var2
    
    def calculate_power(self):
        return self.var1 ** self.var2
    
    def calculate_sum(self, var3):
        return self.var1 + self.var2 + var3
    
calc = Calculator(2, 3)
assert calc.calculate_power() == 8
assert calc.calculate_sum(4) == 9
```

```python
'''
    2. Завершите класс StringManipulator
       Заполните ____ пропуски и создайте реализацию stripped_title().
'''

class StringManipulator:
    """Docstring of StringManipulator"""
    
    category = 'Manipulator'
    
    def __init__(self, original):
        self.string = original
        
    def reverse_words(self):
        words = self.string.split(' ')
        self.string = ' '.join(reversed(words))
       
        
    def make_title(self):
        self.string = self.string.title()
        
    def get_manipulated(self):
        return self.string
    
assert StringManipulator.__doc__ == 'Docstring of StringManipulator'
assert StringManipulator.category == 'Manipulator'

str_manip = StringManipulator('cOOL pyThON')

str_manip.reverse_words()
assert str_manip.get_manipulated() == 'pyThON cOOL'

str_manip.make_title()
assert str_manip.get_manipulated() == 'Python Cool'
```

```python
'''
    3. Создайте класс Dog
       Создайте класс Dog согласно следующей спецификации:
         1) Dogs расходуют свою энергию при лае и восстанавливают во сне
         2) Новые Dogs имеют 10 ед. энергии
         3) Dog имеет метод sleep который восстанавливает энергию на 2 ед.
         4) Dog имеет метод bark который уменьшает энергию на 1 ед.
         5) Dog имеет метод get_energy который возвращает кол-во оставшейся энергии
'''

class Dog:
    def __init__(self):
        self.energy = 10
            
    def get_energy(self):
        return int(self.energy)
    
    def bark(self):
        if int(self.energy) > 0:
            self.energy = int(self.energy) - 1
    
    def sleep(self):
        if int(self.energy) <=8:
            self.energy = int(self.energy) + 2
    
    
doge = Dog()
assert doge.get_energy() == 10

doge.bark()
doge.bark()
doge.bark()
assert doge.get_energy() == 7

doge.sleep()
assert doge.get_energy() == 9

another_doge = Dog()
assert another_doge.get_energy() == 10
```
